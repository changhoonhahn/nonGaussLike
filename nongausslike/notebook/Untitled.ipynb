```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
a = np.random.normal(size=1000)
```

```python
b = np.random.normal(size=1000)
```

```python
c = np.log10(a/b)
```

```python
pa, be = np.histogram(a, normed=True)
plt.plot(0.5*(be[1:]+be[:-1]), pa)
pb, be = np.histogram(b, normed=True)
plt.plot(0.5*(be[1:]+be[:-1]), pb)
pc, be = np.histogram(c[np.isfinite(c)], normed=True)
plt.plot(0.5*(be[1:]+be[:-1]), pc)
```
